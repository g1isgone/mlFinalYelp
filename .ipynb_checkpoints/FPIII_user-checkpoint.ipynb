```python
'''Return all review_id for df in a list, assuming there are no duplicates'''
def getReviews(df):
#     reviewSet=set()
#     for review in df['Reviews']:
#         reviewSet=reviewSet.union(review)
    reviews=[]
    for reviewset in df['Reviews']:
        for review in reviewset:
            reviews.append(review)
    return reviews
```

## Now go through all review data frames and put all inactive reviews in one data frame, and all active reviews in another data frame

```python
def groupByActivity(csvfile,inactiveReviews,activeReviews):
    reviewDf=pd.read_csv(csvfile,index_col=0)
    inactiveDf=reviewDf.loc[reviewDf['review_id'].isin(inactiveReviews)]
    activeDf=reviewDf.loc[reviewDf['review_id'].isin(activeReviews)]
    return inactiveDf,activeDf
```

```python
inactiveDf1,activeDf1=groupByActivity('reviewDfFile2.csv',inactiveReviews,activeReviews)
```

```python
len(activeDf1)
```

```python
len(inactiveDf1)
```

```python
inactiveDf2,activeDf2=groupByActivity('reviewDfFile2.csv',inactiveReviews,activeReviews)
```

```python
print(len(inactiveDf2))
print(len(activeDf2))
```

```python
inactiveDf3,activeDf3=groupByActivity('reviewDfFile3.csv',inactiveReviews,activeReviews)
```

```python
print(len(inactiveDf3))
print(len(activeDf3))
```

```python
inactiveDf4,activeDf4=groupByActivity('reviewDfFile4.csv',inactiveReviews,activeReviews)
```

```python
print(len(inactiveDf4))
print(len(activeDf4))
```

```python
len(inactiveDf1)+len(inactiveDf2)+len(inactiveDf3)+len(inactiveDf4)==len(inactiveReviews)
```

```python
inactiveDf=pd.concat([inactiveDf1,inactiveDf2,inactiveDf3,inactiveDf4])
```

```python
activeDf=pd.concat([activeDf1,activeDf2,activeDf3,activeDf4])
```
