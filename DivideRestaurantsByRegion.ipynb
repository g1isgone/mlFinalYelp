```python
uniqueStates = restaurantDf['state'].unique()
statesCount = {}
for state in restaurantDf['state']:
    if(state in statesCount):
        statesCount[state] = statesCount[state]+1 
    else:
        statesCount[state] = 1;
```

```python
statesCount
```

```python
importantStatesCount = statesCount.copy()
for mapping in statesCount:
    if(importantStatesCount[mapping] < 200):
        del importantStatesCount[mapping]
```

```python
importantStatesCount
```

```python
def createStateDataFrame(stateName, df, columnName): 
    return df.loc[df[columnName] == stateName].reset_index()
```

```python
arizonaDf = createStateDataFrame('AZ', restaurantDf, 'state')
bwDf = createStateDataFrame('BW', restaurantDf, 'state')
edhDf = createStateDataFrame('EDH', restaurantDf, 'state')
illinoisDf = createStateDataFrame('IL', restaurantDf, 'state')
ncDf = createStateDataFrame('NC', restaurantDf, 'state')
nvDf = createStateDataFrame('NV', restaurantDf, 'state')
ohioDf = createStateDataFrame('OH', restaurantDf, 'state')
onDf = createStateDataFrame('ON', restaurantDf, 'state')
paDf = createStateDataFrame('PA', restaurantDf, 'state')
qcDf = createStateDataFrame('QC', restaurantDf, 'state')
scDf = createStateDataFrame('SC', restaurantDf, 'state')
wisconsinDf = createStateDataFrame('WI', restaurantDf, 'state')
```
